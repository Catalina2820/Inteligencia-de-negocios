```python
import pandas as pd
import numpy as np
from google.colab import files
import io
from tqdm import tqdm
```

```python
uploaded=files.upload()
```

```python
uploaded2=files.upload()
```

```python
data_w=pd.read_csv(io.BytesIO(uploaded2['w2.csv']))
data_r=pd.read_csv(io.BytesIO(uploaded['w1.csv']))
```

```python
data_w.head()
```

```python
#Pongo las etiquetas 0 para blanco/ 1 para rojo
data_w['Tipo']=0
data_r['Tipo']=1
```

```python
len(data_r)
```

```python
data=pd.concat([data_w[0:1000],data_r[0:1000]], axis=0)
```

```python
data
```

```python
data=data.reset_index()
```

```python
data
```

```python
data=data.drop('index', axis=1)
```

```python
data
```

```python
from sklearn import preprocessing, neighbors
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
train, test= train_test_split(data, test_size=0.2)
```

```python
train
```

```python
test
```

```python
X_train=train.iloc[:,[0,1,3,4,5,7,8,10]]
Y_train=train['Tipo']
X_test=test.iloc[:,[0,1,3,4,5,7,8,10]]
Y_test=test['Tipo']
```

```python
#Armar el clasificador
clf=neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
#Entrenar el clasificador
clf.fit(X_train,Y_train)
```

```python
accuracy=clf.score(X_test,Y_test)
```

```python
print(accuracy)
```

```python
#Cargo para trabajar otro clasificador llamado árbol de decisión
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
```

```python
#Armar clasificador
tree=DecisionTreeClassifier(criterion="entropy", min_samples_split=20, random_state=10)
```

```python
#Entrenar el árbol de decisión
tree.fit(X_train,Y_train)
```

```python
for i in tqdm(range(1,11)):
  tree=DecisionTreeClassifier(criterion="entropy",max_depth=i, min_samples_split=20, random_state=10)
  tree.fit(X_train,Y_train)
  cv=KFold(n_splits=10, shuffle=True, random_state=10)
  cv.get_n_splits(X_train)
  scores=cross_val_score(tree,X_train,Y_train, scoring="accuracy", cv=cv, n_jobs=-1)
  score=np.mean(scores)
  print("Resultado para i = ", i, " es de ", score)


```
